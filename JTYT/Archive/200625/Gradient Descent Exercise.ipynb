```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
## Construct true Data Generation Process

x=np.expand_dims(np.arange(1,51,1),axis=1) 
beta=np.array([[7]]) 
randomerror=np.random.normal(size=(50,1))*30

y=20+ beta*x+randomerror

plt.scatter(x,y)   ## when do i need to squeeze ? 
 
```

```python
## Gradient Descent with Iteration 

alpha=0
beta=0
lr=1e-5
iter=100000
iterlist=np.arange(0,iter+1)


def gradientloss(y,y_hat,x):
    return (1/(len(y))* (x.T @ (y_hat-y)))

for i in iterlist:
    y_hat=beta*x+alpha
    beta=beta-lr*gradientloss(y,y_hat,x)
    alpha=alpha-lr*np.sum(y_hat-y)
    
print(alpha)
print(beta)


plt.scatter(x,y)
plt.plot(x,y_hat)
```

```python
## Gradient Descent with Iteration and Precision metric

alpha=0
beta=0
lr=1e-5
maxiters=100000
precision=1e-8
i=0
prevstepbeta=1
prevstepalpha=1

def gradientloss(y,y_hat,x):
    return (1/(len(y))* (x.T @ (y_hat-y)))


while i < maxiters and prevstepbeta>precision and prevstepalpha>precision:
    y_hat=beta*x+alpha
    y_hat_minus1=y_hat
    beta_minus1=beta
    alpha_minus1=alpha
    beta=beta-lr*gradientloss(y,y_hat,x)
    alpha=alpha-lr*np.sum(y_hat-y)
    y_hat=beta*x+alpha
    prevstepbeta = abs(beta-beta_minus1)
    prevstepalpha=abs(alpha-alpha_minus1)
    i=i+1

print(alpha)
print(beta)

plt.scatter(x,y)
plt.plot(x,y_hat)
```
