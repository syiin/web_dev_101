````python
import pandas as pd
import numpy as np
from itertools import permutations 
````

# My Misc Setup

````python
perm1 = list(permutations([1, 2, 3])) 
perm2 = list(permutations([4, 5, 6])) 
perm3 = list(permutations([7, 8, 9])) 

perm = perm1 + perm2 + perm3

def convert_to_str(x):
    return str(x)
perm = [list(o) for o in perm]
perm = list(map(convert_to_str, perm))

df = pd.DataFrame(perm, columns=["list_of_ids"])
````

````python
type(df.iloc[0]['list_of_ids'])
````

# Exercises

1. Given the dataframe here, convert the column of string representations of lists into integers. ie. notice that running `type(df.iloc[0]['list_of_ids'])` returns `str`. We want this to return list for the whole column - it should be the same for `type(df.iloc[1]['list_of_ids'])`, `type(df.iloc[2]['list_of_ids'])`...etc

````python
def convert_to_list(string):
    return eval(string)

df['list_of_ids']=df['list_of_ids'].apply(convert_to_list) 


````

````python
print(type(df.iloc[0]['list_of_ids']))
print(type(df.iloc[0]['list_of_ids'][0]))
````
