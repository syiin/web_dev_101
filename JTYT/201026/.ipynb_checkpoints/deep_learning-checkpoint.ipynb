```python
import matplotlib.pyplot as plt
from sklearn import datasets

plt.style.use('dark_background')
```

# Using plt.subplot

```python
#create a Figure instance that would contain all the plot elements and set the figure size
plt.figure(figsize=(10,5))

ax1 = plt.subplot(1,2,1)# number of row = 1, number of columns = 2, subplot order = 1 
# OR can be written as ax1 = plt.subplot(121)

ax1.imshow(digits.images[0])

ax1.set_title('ZERO') 
ax1.set_xlabel('width')
ax1.set_ylabel('height')

ax2 = plt.subplot(1,2,2)
ax2.imshow(digits.images[1]) # this is where you put your plot if any (eg. ax2.scatter())
ax2.set_title('ONE') 
ax2.set_xlabel('width')
ax2.set_ylabel('height')

plt.subplots_adjust(wspace=0.4)
plt.show()
```

For loop

```python
plt.figure(figsize=(10,10))
nrows = 3
ncols = 3
for i in range(nrows * ncols):
    ax = plt.subplot(nrows, ncols, i + 1)
    ax.imshow(digits.images[i])
    
    ax.set_title(f"NUM: {digits.target[i]}") 
    ax.set_xlabel('width')
    ax.set_ylabel('height')
    
    plt.subplots_adjust(top = 1.0 , bottom=0.05) #padding between subplots
```

```python
digits.target[i]
```

# Using plt.subplots using axes

```python
fig,axes = plt.subplots(nrows =1, ncols =2, figsize =(10,5))
ax1 = axes[0]

ax1.imshow(digits.images[0])
ax1.set_title(f"NUM: {digits.target[0]}") 
ax1.set_xlabel('width')
ax1.set_ylabel('height')

ax2 = axes[1]
ax2.imshow(digits.images[1])
ax2.set_title(f"NUM: {digits.target[1]}") 
ax2.set_xlabel('width')
ax2.set_ylabel('height')

plt.subplots_adjust(wspace=0.4)
plt.show()
```

```python
fig,axes = plt.subplots(nrows =3, ncols =3, figsize =(10,10))
for i,row in enumerate(axes):
    for j,cell in enumerate(row): 
        idx = i * j
        cell.imshow(digits.images[idx])
        cell.set_title(f"NUM: {digits.target[idx]}") 
        cell.set_xlabel('width')
        cell.set_ylabel('height')
plt.subplots_adjust(hspace=0.5)
```

```python
axes.shape
```

# Using plt.figure.add_subplot

```python
fig = plt.figure(figsize=(10,5))

ax1 = fig.add_subplot(1,2,1) #NOT plt.subplot(3,3)
ax1.imshow(digits.images[0]) #or if a dataframe, df.Series.plot()
ax1.set_title('ZERO') 
ax1.set_xlabel('width')
ax1.set_ylabel('height')

ax2 = fig.add_subplot(1,2,2)
ax2.imshow(digits.images[1])
ax2.set_title('ONE') 
ax2.set_xlabel('width')
ax2.set_ylabel('height')

plt.subplots_adjust(wspace=0.4)
plt.show()
```

```python
fig = plt.figure(figsize=(10,10))
nrows = 3
ncols = 3
for i in range(nrows * ncols):
    ax = fig.add_subplot(nrows, ncols, i + 1)
    ax.imshow(digits.images[i])
    
    ax.set_title(f"NUM: {digits.target[i]}") 
    ax.set_xlabel('width')
    ax.set_ylabel('height')
    
    plt.subplots_adjust(top = 1.0 , bottom=0.05) #padding between subplots

plt.show()
```
